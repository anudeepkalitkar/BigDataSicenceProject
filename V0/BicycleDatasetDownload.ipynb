```python
def generate_time_sequence(start_datetime):
    current_datetime = datetime.now()
    current_datetime = current_datetime.replace(second=0, microsecond=0)

    time_sequence = []

    while start_datetime <= current_datetime:
        time_sequence.append(start_datetime)
        start_datetime += timedelta(minutes=15)

    return time_sequence

# Example usage:
start_datetime = datetime(2023, 11, 1, 0, 0)  # Replace this with your desired start datetime
time_sequence = generate_time_sequence(start_datetime)

timeInterval = []
for time_point in time_sequence:
    # print(time_point.strftime("%Y-%m-%d %H:%M"))
    timeInterval.append(time_point.strftime("%Y-%m-%d %H:%M"))
print(timeInterval)
```
